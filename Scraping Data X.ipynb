```python
# Menyimpan nama file output hasil scraping tweet ke dalam variabel 'filename'
filename = 'naturalisasi.csv'

# Menyimpan kata kunci pencarian untuk tweet ke dalam variabel 'search_keyword'
# Pencarian tweet yang mengandung "erick thohir", dari 1 Januari 2024 sampai 31 Desember 2024, berbahasa Indonesia
search_keyword = 'naturalisasi timnas since:2024-01-01 until:2024-12-31 lang:id'

# Menyimpan batas jumlah tweet yang ingin diambil (maksimal 2000 tweet)
limit = 2000

# Menjalankan tool tweet-harvest versi 2.6.1 menggunakan npx untuk scraping tweet
# -o "{filename}" → nama file output (csv)
# -s "{search_keyword}" → kata kunci pencarian tweet
# --tab "LATEST" → mengambil tweet dari tab 'Terbaru' (LATEST) di Twitter
# -l {limit} → jumlah maksimal tweet yang diambil
# --token {twitter_auth_token} → menggunakan token autentikasi Twitter yang sudah disimpan sebelumnya
!npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}
```

```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")

# Display the DataFrame
display(df)
```
